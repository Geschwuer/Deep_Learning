```python
import numpy as np
import matplotlib.pyplot as plt

# Wertebereich für X und Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Gitterpunkte erzeugen
X, Y = np.meshgrid(x, y)

# Funktion auf Gitter anwenden
Z = X**2 + Y**2

# 3D-Plot erzeugen
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Achsenbeschriftungen und Titel
ax.set_xlabel('X-Achse')
ax.set_ylabel('Y-Achse')
ax.set_zlabel('Z = X² + Y²')
ax.set_title('3D-Oberfläche: Z = X² + Y²')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Auflösung
resolution_x = 10
resolution_y = 10

# Gitter erzeugen
x = np.linspace(0, 1, resolution_x)
y = np.linspace(0, 1, resolution_y)
X, Y = np.meshgrid(x, y)

# Farbverlauf ähnlich dem Bild
R = X                    # Rot nimmt von links nach rechts zu
G = Y                    # Grün nimmt von oben nach unten zu
B = 1 - X                # Blau nimmt von rechts nach links ab

# Kombinieren zu RGB
RGB_image = np.stack([R, G, B], axis=2)

# Anzeigen
plt.imshow(RGB_image)
plt.title("Ähnlicher Farbverlauf wie im Bild")
plt.axis('off')
plt.show()
```
